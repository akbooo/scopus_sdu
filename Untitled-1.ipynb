```python
import pandas as pd
import re
import unicodedata

# ----------------------------
# 1. LOAD DATA
# ----------------------------
df = pd.read_csv("рез2.csv")   # <-- ваш основной CSV
teachers = pd.read_csv("teachers.csv", encoding="utf-8-sig")
aliases = pd.read_csv("aliase.csv", encoding="utf-8-sig", sep=";")  # если есть

df.columns = [c.strip() for c in df.columns]

df = df.rename(columns={
    "Автор (ы)": "authors",
    "Author full names": "author_full_names",
    "Идентификатор автора(ов)": "author_ids",
    "Название документа": "title",
    "Год": "year",
    "Название источника": "journal",
    "Цитирования": "citations",
    "DOI": "doi",
    "Ссылка": "scopus_link",
    "Организации": "affiliations",
    "Тип документа": "doc_type",
    "Open Access (открытый доступ)": "open_access",
    "EID": "eid",
    "subject_area": "subject_area",
    "Квартиль_лучший": "quartile",
    "Квартили_все": "quartile_all",
    "Квартили_детально": "quartile_detail",
    "API_Status": "quartile_status",
})

df["citations"] = pd.to_numeric(df["citations"], errors="coerce").fillna(0).astype(int)
df["year"] = pd.to_numeric(df["year"], errors="coerce")

# ----------------------------
# 2. HELPERS FROM DASHBOARD
# ----------------------------
def _normalize_str(s):
    cyr_map = {
        'А':'A','Б':'B','В':'V','Г':'G','Д':'D','Е':'E','Ё':'E','Ж':'Zh',
        'З':'Z','И':'I','Й':'Y','К':'K','Л':'L','М':'M','Н':'N','О':'O',
        'П':'P','Р':'R','С':'S','Т':'T','У':'U','Ф':'F','Х':'Kh','Ц':'Ts',
        'Ч':'Ch','Ш':'Sh','Щ':'Sch','Ъ':'','Ы':'Y','Ь':'','Э':'E','Ю':'Yu',
        'Я':'Ya',
        'а':'a','б':'b','в':'v','г':'g','д':'d','е':'e','ё':'e','ж':'zh',
        'з':'z','и':'i','й':'y','к':'k','л':'l','м':'m','н':'n','о':'o',
        'п':'p','р':'r','с':'s','т':'t','у':'u','ф':'f','х':'kh','ц':'ts',
        'ч':'ch','ш':'sh','щ':'sch','ъ':'','ы':'y','ь':'','э':'e','ю':'yu',
        'я':'ya',
    }
    result = ''.join(cyr_map.get(ch, ch) for ch in str(s))
    result = result.lower().strip()
    result = unicodedata.normalize('NFD', result)
    result = ''.join(c for c in result if unicodedata.category(c) != 'Mn')
    return result

def _levenshtein(a, b):
    if a == b:
        return 0
    if len(a) < len(b):
        a, b = b, a
    prev = list(range(len(b) + 1))
    for i, ca in enumerate(a):
        curr = [i + 1]
        for j, cb in enumerate(b):
            curr.append(min(prev[j] + (ca != cb), curr[j] + 1, prev[j + 1] + 1))
        prev = curr
    return prev[-1]

def build_fuzzy_map(mapping, max_dist=2):
    by_initial = {}
    surname_to_staff = {}
    for k in mapping:
        if ',' not in k:
            continue
        surname, rest = k.split(',', 1)
        initial = rest.strip()[:1]
        by_initial.setdefault(initial, []).append((surname.strip(), k))
        surname_to_staff.setdefault(surname.strip(), k)
    return {"by_initial": by_initial, "surname_to_staff": surname_to_staff}

def _resolve_key(key, author_dept, fuzzy_map, max_dist=2):
    if key in author_dept:
        return key
    if ',' not in key:
        return None
    surname_q, rest_q = key.split(',', 1)
    surname_q = surname_q.strip()
    initial_q = rest_q.strip()[:1]

    best_key, best_dist = None, max_dist + 1
    for surname_c, staff_key in fuzzy_map["by_initial"].get(initial_q, []):
        if abs(len(surname_q) - len(surname_c)) > max_dist:
            continue
        d = _levenshtein(surname_q, surname_c)
        if d <= max_dist and d < best_dist:
            best_dist = d
            best_key = staff_key
    return best_key

def build_author_dept_map(surnames, names, staff_types, statuses, departments, emp_ids=None):
    mapping = {}
    if emp_ids is None:
        emp_ids = [None] * len(surnames)

    for surname, name, stype, status, dept, emp_id in zip(
        surnames, names, staff_types, statuses, departments, emp_ids
    ):
        if stype != "Academic":
            continue
        if pd.isna(surname) or pd.isna(name) or not str(name).strip():
            continue

        norm_surname = _normalize_str(str(surname).strip())
        norm_name = _normalize_str(str(name).strip())
        first_initial = next((c for c in norm_name if c.isalpha()), "")
        if not first_initial:
            continue

        info = {"dept": str(dept or ""), "status": str(status or "")}

        key_init = f"{norm_surname}, {first_initial}."
        key_full = f"{norm_surname}, {norm_name}"

        if key_init not in mapping:
            mapping[key_init] = info
        if key_full not in mapping:
            mapping[key_full] = info

        if pd.notna(emp_id) and str(emp_id).strip() not in ("", "nan", "None"):
            mapping[f"empid:{str(emp_id).strip()}"] = info

    return mapping

def normalize_pub_author(raw_author):
    raw = str(raw_author).strip()
    raw = re.sub(r'\(\d+\)', '', raw).strip()
    if not raw:
        return []

    if ',' in raw:
        parts = raw.split(',', 1)
        surname = _normalize_str(parts[0].strip())
        rest = _normalize_str(parts[1].strip())
    else:
        tokens = raw.split()
        if len(tokens) < 2:
            return []
        surname = _normalize_str(tokens[-1].strip())
        rest = _normalize_str(tokens[0].strip())

    first_initial = next((c for c in rest if c.isalpha()), '')
    if not first_initial:
        return []

    keys = []
    rest_clean = rest.replace('.', '').strip()
    if len(rest_clean) > 2:
        keys.append(f"{surname}, {rest_clean}")
    keys.append(f"{surname}, {first_initial}.")
    return keys

def parse_full_authors(s):
    if pd.isna(s) or not str(s).strip():
        return []
    result = []
    for entry in str(s).split(";"):
        entry = entry.strip()
        id_match = re.search(r'\((\d+)\)', entry)
        scopus_id = id_match.group(1) if id_match else None
        entry = re.sub(r'\(\d+\)', '', entry).strip()
        if ',' in entry:
            parts = entry.split(',', 1)
            lastname = parts[0].strip()
            firstname = parts[1].strip()
            if lastname and firstname:
                result.append((lastname, firstname, scopus_id))
    return result

df["author_full_list"] = (
    df["author_full_names"].apply(parse_full_authors)
    if "author_full_names" in df.columns
    else [[] for _ in range(len(df))]
)

# ----------------------------
# 3. BUILD FACULTY MATCHING
# ----------------------------
emp_ids = teachers["EMP_ID"].tolist() if "EMP_ID" in teachers.columns else None

author_dept = build_author_dept_map(
    teachers["SURNAME"].tolist(),
    teachers["NAME"].tolist(),
    teachers["STAFF_TYPE_EN"].tolist(),
    teachers["STATUS_EN"].tolist(),
    teachers["DEPARTMENT_EN"].tolist(),
    emp_ids=emp_ids,
)

if not aliases.empty and {"scopus_name", "emp_id"}.issubset(set(aliases.columns.str.lower())):
    aliases.columns = [c.strip().lower() for c in aliases.columns]
    for _, row in aliases.iterrows():
        name = str(row["scopus_name"]).strip()
        eid = str(row["emp_id"]).strip()
        if not name or not eid or name in ("nan", "None"):
            continue

        if "," in name:
            parts = name.split(",", 1)
            scopus_norm_key = _normalize_str(parts[0].strip()) + ", " + _normalize_str(parts[1].strip())
        else:
            scopus_norm_key = _normalize_str(name)

        empid_key = f"empid:{eid}"
        if empid_key in author_dept:
            author_dept[scopus_norm_key] = author_dept[empid_key]

            if "," in scopus_norm_key:
                parts = scopus_norm_key.split(",", 1)
                sur = parts[0].strip()
                ini = next((ch for ch in parts[1].strip() if ch.isalpha()), "")
                if ini:
                    init_key = f"{sur}, {ini}."
                    if init_key not in author_dept:
                        author_dept[init_key] = author_dept[empid_key]

fuzzy_map = build_fuzzy_map(author_dept)

# ----------------------------
# 4. MATCH PUBLICATIONS EXACTLY LIKE FACULTY TAB
# ----------------------------
pub_dept_rows = []
unmatched_rows = []

for idx, row in df.iterrows():
    depts_seen = set()
    matched_any = False
    matched_keys = []

    full_authors = row.get("author_full_list", [])
    if not isinstance(full_authors, list):
        full_authors = []

    authors_raw = row.get("authors", "")
    short_authors = str(authors_raw).split(";") if pd.notna(authors_raw) else []

    candidates = []
    for lastname, firstname, *_ in full_authors:
        candidates.append(f"{_normalize_str(lastname)}, {_normalize_str(firstname)}")
    for a in short_authors:
        for k in normalize_pub_author(a.strip()):
            candidates.append(k)

    for key in candidates:
        resolved = _resolve_key(key, author_dept, fuzzy_map)
        info = author_dept.get(resolved) if resolved else None
        if not info:
            continue

        matched_any = True
        matched_keys.append(resolved)

        dept = info["dept"]
        staff_status = info["status"]

        if dept and dept not in depts_seen:
            depts_seen.add(dept)
            pub_dept_rows.append({
                "pub_idx": idx,
                "department": dept,
                "staff_status": staff_status,
                "citations": int(row.get("citations", 0) or 0),
            })

    if not matched_any:
        unmatched_rows.append({
            "pub_idx": idx,
            "title": row.get("title"),
            "authors": row.get("authors"),
            "year": row.get("year"),
            "citations": row.get("citations"),
            "doi": row.get("doi"),
            "affiliations": row.get("affiliations"),
            "eid": row.get("eid"),
        })

pub_dept_df = pd.DataFrame(pub_dept_rows)
matched_idx = set(pub_dept_df["pub_idx"].unique()) if not pub_dept_df.empty else set()

faculty_view = df[df.index.isin(matched_idx)].copy()
not_in_faculty = df[~df.index.isin(matched_idx)].copy()

print("ALL citations:", df["citations"].sum())
print("FACULTY citations:", faculty_view["citations"].sum())
print("DIFF:", df["citations"].sum() - faculty_view["citations"].sum())

# ----------------------------
# 5. ONLY Cited missing papers
# ----------------------------
missing_cited = not_in_faculty[not_in_faculty["citations"] > 0].copy()

cols = ["title", "authors", "year", "citations", "doi", "journal", "affiliations", "eid"]
missing_cited = missing_cited[cols].sort_values("citations", ascending=False)

print("\nMissing cited papers:")
print(missing_cited.to_string(index=False))

print("\nMissing cited papers total citations:", missing_cited["citations"].sum())

missing_cited.to_csv("missing_from_faculty_exact.csv", index=False, encoding="utf-8-sig")
print("\nSaved: missing_from_faculty_exact.csv")
```
