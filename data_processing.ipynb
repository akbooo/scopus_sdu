```python
publications = (df.groupby("EID", as_index=False)
                  .agg(
                      title=("Название документа", "first"),
                      year=("Год", "first"),
                      citations=("Цитирования", "first"),
                      subject_areas=("subject_area", lambda s: sorted(set(s)))
                  ))

# чтобы смотреть как строку (удобно для UI/CSV)
publications["subject_areas_str"] = publications["subject_areas"].apply(lambda x: "; ".join(x))

# проверка
print(any("health_professions" in x for x in publications["subject_areas"]))
```
